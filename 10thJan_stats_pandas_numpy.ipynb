```python
from statsmodels.stats.weightstats import ztest
pop_mean = 70
sig_level = 0.05 #threshold

print("you are going to enter the samples data now")
sample_scores1 = []
for i in range(1, 6):
  s = int(input(f"Enter sample1\'s match no {i}\'s score:"))
  sample_scores1.append(s)

sample_scores2 = []
for i in range(1, 6):
  sample_scores2.append(int(input(f"Enter sample2's match no {i}\' score:")))

sample_scores3 = []
for i in range(1, 6):
  sample_scores3.append(int(input(f"Enter sample3's match no {i}\' score:")))

all_samples = [sample_scores1, sample_scores2, sample_scores3]

lst_pvalues = []

for i in all_samples:
  zvaluei, pvaluei = ztest(i, value = pop_mean)
  lst_pvalues.append(pvaluei)

print(lst_pvalues)

counterH0 = 0
counterH1 = 0

for i in lst_pvalues:
  if (i>sig_level):
    counterH0 += 1
    print(f"Null Hypothesis is to be accepted")
  else:
    counterH1 += 1
    print(f"Null Hypothesis is to be rejected. H1 is to be accepted.")


if(counterH0 > counterH1):
  print(f"Finally H0 is proved ..\"and my friends claim is true\" with {int(counterH0 * 100)/3} % confidance.")
else:
  print(f"Finally H1 is proved ..\"and my friends claim is wrong\" with {int(counterH1 * 100)/3} % confidance.")
```

```python
#method 4 => statistically researched method
def remove_outlier(df_in, col_name):
  q1 = df_in[col_name].quantile(0.25)
  q3 = df_in[col_name].quantile(0.75)
  iqr = q3 - q1
  lower_whisker = q1 - 1.5 * iqr
  upper_whisker = q3 + 1.5 * iqr
  df_out = df_in.loc[(df_in[col_name]< upper_whisker) & (df_in[col_name] > lower_whisker)]
  return df_out
```

```python
#outliers lies on both of the ends of data
def remove_outliers_US(df, col_name, persent):
  p1 = float(persent)/200.0
  p2 = 1 - p1
  lower = df[col_name].quantile(p1)
  upper = df[col_name].quantile(p2)
  df_out = df[(df[col_name] < upper) & (df[col_name] > lower)]
  return df_out
```

```python
#outliers are in lower range

def remove_outliers_US(df, col_name, persent):
  p = float(persent)/100.0
  crirical = df[col_name].quantile(p)
  df_out = df[(df[col_name] > crirical)]
  return df_out
```

```python
#outliers are in upper range

def remove_outliers_US(df, col_name, persent):
  p = float(persent)/100.0
  crirical = df[col_name].quantile(1-p)
  df_out = df[(df[col_name] > crirical)]
  return df_out
```

```python
# Example : Create a filter array that will return only values higher than 42:
import numpy as np

arr = np.array([41, 42, 43, 44])

# Create an empty list
filter_arr = []

# go through each element in arr
for element in arr:
    # if the element is higher than 42, set the value to True, otherwise False:
    if element > 42:
        filter_arr.append(True)
    else:
        filter_arr.append(False)

newarr = arr[filter_arr]

print(filter_arr)
print(newarr)
```

```python
# Example : Create a filter array that will return only values higher than 42:
import numpy as np
arr = np.array([41, 42, 43, 44])

filter_arr = arr > 42

newarr = arr[filter_arr]

print(filter_arr)
print(newarr)
```

```python
# Example : Create a filter array that will return only odd elements from the original array:
import numpy as np
arr = np.array([1, 2, 3, 4, 5, 6, 7])

filter_arr = arr % 2 != 0

newarr = arr[filter_arr]

print(filter_arr)
print(newarr)
```
